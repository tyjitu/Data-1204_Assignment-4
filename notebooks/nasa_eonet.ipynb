```python
import requests, json
from datetime import datetime
from pathlib import Path

project_root = Path.cwd().parent

# API endpoint
url = "https://eonet.gsfc.nasa.gov/api/v3/events"

# Parameters for API request
params = {
    "status": "all",
    "start": "2024-01-01",
    "end": "2024-12-31",
    "category": "wildfires,severeStorms"
}

try:
    # Send GET request
    response = requests.get(url, params=params, timeout=30)
    response.raise_for_status()
    data = response.json()

    # Create timestamped filename
    ts = datetime.now().strftime("%Y-%m-%dT%H-%M-%S")
    filename = project_root / "data" / "bronze" / "nasa_eonet" / f"events_{ts}.json"

    # Create directories if they don't exist
    filename.parent.mkdir(parents=True, exist_ok=True)

    # Save JSON response
    with open(filename, "w") as f:
        json.dump(data, f, indent=2)

    print(f"Saved {filename}")

except requests.exceptions.Timeout:
    print("Request timed out. Check your internet connection or try again later.")
except requests.exceptions.RequestException as exc:
    print(f"NASA EONET request failed: {exc}")
```
